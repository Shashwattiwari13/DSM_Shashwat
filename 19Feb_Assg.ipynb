Q1. What is Flask Framework? What are the advantages of Flask Framework?

Ans:
Flask is a python Micro-framework which is used to build web applications.Flask relies on the WSGI(Web Server Gateway Interface) external library to function, as well as the Jinja2 template engine.
It provides us with various libraries,modules and structure which we used to build web applications.

Advantages:
Scalable:

If you want to make an app which starts small but has a potiential to grow big.

Flexibility:

The minimal nature of Flask and its aptitude for developing smaller web apps means that it’s even more flexible than Django itself.


Easy to negotiate:

The microframework is easy to understand for web developers, not just saving them time and effort but also giving them more control over their code and what is possible.

Q3. What is App routing in Flask? Why do we use app routes?

Ans:

App routing in flask means mapping the function to an url so that the it's logic could be handled.
We use app routing because it includes the use of url which is langauge independent to access a function.
The decorator @app.route('url') is specified above the function so as to map the function to the url

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

Ans:

url_for() function is used for URL Building in flask.

It accepts the name of the function as its first argument and any number of keyword arguments, each corresponding to a variable part of the URL rule.

```python
from flask import *

app2=Flask(__name__)

@app2.route('/admin')
def admin_func():
    return "hello admin"

@app2.route('/guest/<guest>')
def guest_func(guest):
    return "hello guest"

@app2.route('/name/<name>')
def user(name):
    if name=='admin':
        return redirect(url_for('admin_func'))
    else:
        return redirect(url_for('guest_func',guest=name))


if __name__=='__main__':
    app2.run(host='0.0.0.0')


```
